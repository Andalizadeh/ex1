```python
# Q1: Create a dataframe which contains information of a company's products:
df = pd.DataFrame({'prod_name'  : ['g1', 'g2', 'c1', 'c2','c3','f1','f2'],
                   'price': [45, 25, 75, 35,89,20,55],
                   'monthly_demand' : [1200,2500,200,1850,120,3500,545],
                   'brand': ['z','x','x','y','z','y','z']})
df
```

```python
df.index= ['g1', 'g2', 'c1', 'c2','c3','f1','f2']
df
```

```python
# a: Add a column to calculate monthly revenue of each product
df['monthly_revenue']=df['price']*df['monthly_demand']
df
```

```python
# b: Calculate total monthly revenue generated by brand z
# df['brand']=='z'
np.sum(df['monthly_revenue'][df['brand']=='z'])
```

```python
#  c: Create a column called revenue share which includes monthly revenue share of each product
df.loc[:,['prod_name','monthly_demand']]
```

```python
#  d: Which brand does have the largest revenue share?
md_mx=df['monthly_demand'].max()
df.loc[df['monthly_demand'].idxmax(), 'brand']
```

```python
#    a: Extract data of married customers with income above average.
# mean_income=data['income'].mean()
# mean_income
data.loc[(data['marital_status'] == 'Married') & (data['income'] > data['income'].mean()), :]
```

```python
#    b: What percentag of female customers are below 26?
round(np.sum((data['sex'] == 'F') & (data['age'] < 26)) / np.sum(data['sex'] == 'F') * 100, 2)
```

```python
#    c: What percentage of female customers above 30 are employed?
round(np.sum((data['sex'] == 'F') & (data['age'] > 30) & (data['is_employed']==True)) / np.sum(data['sex'] == 'F') * 100, 2)
```

```python
#    d: What percentage of cutomers who have health insurance are employed with income above average income?
round(np.sum((data['health_insurance'] == True) & (data['income'] > data['income'].mean()) & (data['is_employed']==True)) / np.sum(data['health_insurance'] == True) * 100, 2)
```

```python
#    e: What percentage of cutomers who have health insurance are married     males?
round(np.sum((data['sex'] == 'M') & (data['health_insurance'] == True ) & (data['marital_status']== 'Married')) / np.sum(data['sex'] == 'F') * 100, 2)
```

```python
#    f: What percentage of married customers have health insurance?
round(np.sum((data['health_insurance'] == True) & (data['marital_status'] == 'Married' ) ) / np.sum(data['marital_status'] == 'Married') * 100, 2)
```
